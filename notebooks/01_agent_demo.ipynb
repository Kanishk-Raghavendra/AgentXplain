# AgentXplain Agent Demo
Run the tool-routing agent on sample prompts and inspect structured outputs.

```python
from pathlib import Path
import sys

root = Path.cwd().resolve().parents[0]
if str(root) not in sys.path:
    sys.path.insert(0, str(root))

from src.agent.agent import AgentXplainAgent
```

```python
agent = AgentXplainAgent(use_mock_router=True)
queries = [
    "calculate 23 + 19",
    "who discovered penicillin",
    "execute python code to print hello",
    "latest news about mars mission",
    "write a function and run code"
]
```

```python
rows = []
for q in queries:
    trace = agent.run(q)
    rows.append({
        "query": trace.query,
        "selected_tool": trace.selected_tool,
        "args": trace.args,
        "reason": trace.reason,
        "tool_score_distribution": trace.tool_score_distribution
    })
rows
```
